```python
import os
import re
import numpy as np

def extract_powfs_numbers(parent_dir, actuator_counts):
    """
    Recursively traverse directories under the given parent directory,
    go into specific subdirectories defined by actuator counts (1000, 1250, ...),
    find files named 'maos_done.conf', extract the first number from lines
    starting with 'powfs.siglev', and return a list of these numbers.

    :param parent_dir: The path to the parent directory to search under.
    :return: A list of extracted numbers.
    """
    powfs_numbers = []

    # Generate actuator directory names using np.arange
    actuator_dirs = [f"{count}_actuators" for count in actuator_counts]

    for actuator_dir in actuator_dirs:
        # Construct the path to the actuator directory
        actuator_path = os.path.join(parent_dir, actuator_dir)

        if os.path.isdir(actuator_path):
            # Construct the path to the 'maos_done.conf' file
            maos_done_path = os.path.join(actuator_path, "maos_done.conf")

            if os.path.isfile(maos_done_path):
                try:
                    with open(maos_done_path, 'r') as f:
                        for line in f:
                            # Match lines containing 'powfs.siglev' and extract the first number
                            match = re.match(r"^#?powfs\.siglev\s*=\s*\[(\d+\.\d+)[^\]]*\]", line)
                            if match:
                                # Extract the first number (group 1)
                                extracted_number = float(match.group(1))
                                powfs_numbers.append(extracted_number)
                                break  # Stop reading after finding the line
                except Exception as e:
                    print(f"Error reading file {maos_done_path}: {e}")

    return powfs_numbers

# Example usage
if __name__ == "__main__":
    parent_directory = act_study  # Replace with your directory path
    actuator_counts = np.arange(1000, 6000, 250)
    result = extract_powfs_numbers(parent_directory, actuator_counts)
    print("Extracted numbers:", result)
```
