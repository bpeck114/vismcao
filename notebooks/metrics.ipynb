```python
# IMPORT NECESSARY PACKAGES
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

# PAARTI
from paarti.utils import maos_utils

# VISMCAO
sys.path.append("../scripts/plotting")
import utils
```

```python
act_study = "../studies/act_count/8_lgs/8mag_lgs/"
```

```python
actuator_count = np.arange(1000, 5776, 250)
seed=1
wfe = np.zeros((3, len(actuator_count)))
```

```python
for i, act_count in enumerate(actuator_count):
    directory_path = os.path.join(act_study, f"{act_count}_actuators/")
    
    if act_count % 1000 == 0:
            print("Entering directory: ", act_count)
    #if act_count == 3250:
       # wfe[0, i] = 145.95693807
       # wfe[1, i] = 69.73466366
       # wfe[2, i] = 128.22053055
       # continue

    wfe_metrics_kola = utils.get_wfe_metrics_over_field(directory=directory_path, seed=seed)
        
    _, clos_mean_nm, *_ = wfe_metrics_kola
           
    wfe[0, i] = clos_mean_nm[0]
    wfe[1, i] = clos_mean_nm[1]
    wfe[2, i] = clos_mean_nm[2]
```

```python
wfe_colors = ["r", "g", "b"]
wfe_labels = ["Total WFE", "Tip-Tilt", "High-Order"]

for i in range(3):
    plt.plot(actuator_count, wfe[i], label=wfe_labels[i], color=wfe_colors[i])

plt.title('Wavefront Error vs. Actuator Count (10 LGS @8mag @30" radius)')
plt.xlabel("Actuator Count (n)")
plt.ylabel("Wavefront Error (nm)")
plt.legend()

plt.show()
```
