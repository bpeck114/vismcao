```python
import os
import subprocess
import numpy as np

# Base paths
# base_path_1 = "../studies/reduced_ngs/9_lgs/8mag_lgs"
# base_path_2 = "../studies/reduced_ngs_broken/8_lgs/8mag_lgs"
base_path_1 = "../master_files/reduced_ngs/8_lgs/8mag_lgs"
base_path_2 = "../master_files/reduced_ngs_with_misreg/8_lgs/8mag_lgs"

# Range of values
counts = np.arange(1000, 6000, 250)

# Loop through the counts and perform diff
for count in counts:
    # file1 = os.path.join(base_path_1, f"{count}count/maos_done.conf")
    # file2 = os.path.join(base_path_2, f"{count}count/maos_done.conf")
    file1 = os.path.join(base_path_1, f"kola_{count}act.conf")
    file2 = os.path.join(base_path_2, f"kola_{count}act.conf")
    
    
    print(f"Comparing: {file1} vs {file2}")
    
    # Run diff command
    try:
        result = subprocess.run(["diff", file1, file2], check=True, text=True, capture_output=True)
        if result.stdout:
            print(result.stdout)
        else:
            print("No differences found.")
    except subprocess.CalledProcessError as e:
        print("Differences found:")
        print(e.stdout)  # Output the differences
    except FileNotFoundError:
        print(f"One of the files is missing: {file1} or {file2}")

    print("-" * 40)
```
